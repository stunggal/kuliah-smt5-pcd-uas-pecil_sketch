```python
img = cv2.imread('../input/sketpensil/944104.png', 1)
```

```python
cv2.imwrite("img_asli.jpg", img)
plt.imshow(cv2.imread('img_asli.jpg'))
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite("hasil_gray.jpg", img_gray)
plt.imshow(cv2.imread('hasil_gray.jpg'))
```

```python
img_invert = cv2.bitwise_not(img_gray)
cv2.imwrite("hasil_invert.jpg", img_invert)
plt.imshow(cv2.imread('hasil_invert.jpg'))
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
cv2.imwrite("hasil_pra.jpg", img_smoothing)
plt.imshow(cv2.imread('hasil_pra.jpg'))
```

```python
final_img = cv2.divide(img_gray, 255 - img_smoothing, scale=256)
cv2.imwrite("hasil.jpg", final_img)
plt.imshow(cv2.imread('hasil.jpg'))
```
